```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from keras import layers
from keras.preprocessing.image import ImageDataGenerator
```

# Importing Data

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    zoom_range=0.2,
    fill_mode='nearest'
)
test_datagen = ImageDataGenerator(rescale=1/255)
validation_datagen = ImageDataGenerator(rescale=1/255)

train_generator = train_datagen.flow_from_directory(
        './train/', 
        batch_size = 1,
        target_size=(100, 100),
        class_mode='binary')

test_generator = test_datagen.flow_from_directory(
        './test/',
        batch_size = 1,
        target_size=(100, 100),
        class_mode='binary')

validation = validation_datagen.flow_from_directory(
    './val/',
    batch_size = 1,
    class_mode='binary',
    target_size=(100, 100))
```

# Data visualization

```python
x_train = []
y_train = []
x_val = []
y_val = []
x_test = []
y_test = []

c = 0
for feature, label in train_generator:
    x_train.append(np.array(feature))
    y_train.append(np.array(label))
    c += 1
    if c >= 16:
        break

c = 0
for feature, label in test_generator:
    x_test.append(np.array(feature))
    y_test.append(np.array(label))
    c += 1
    if c >= 16:
        break

c = 0
for feature, label in validation:
    x_val.append(np.array(feature))
    y_val.append(np.array(label))
    c += 1
    if c >= 16:
        break

x_train = np.array(x_train)
print(x_train.shape)
x_train = np.reshape(x_train, (16, 100, 100, 3))
print(x_train.shape)
```

```python
x_val = np.array(x_val)
print(x_val.shape)
x_val = np.reshape(x_val, (16, 100, 100, 3))
print(x_val.shape)
```

```python
fig, ax = plt.subplots(4, 4, figsize=(30, 30))
for i in range(0, 16):
    ax[int(i / 4), (i % 4)].imshow(x_train[i])
```

# Model building

```python
model = tf.keras.models.Sequential([
    # This is the first convolution
    layers.Conv2D(16, (3,3), activation='relu', input_shape=(100, 100, 3)),
    layers.MaxPooling2D(2, 2),
    # The second convolution
    layers.Conv2D(32, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),
    # The third convolution
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),
    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.Dropout(0.15),
    layers.Dense(1, activation='sigmoid')
])
```

# Model compiling and training

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
history = model.fit(
      train_generator, 
      epochs=15,
      verbose=1,
      validation_data = test_generator)
```

# Model performance visualization

```python
acc=history.history['accuracy']
val_acc=history.history['val_accuracy']
loss=history.history['loss']
val_loss=history.history['val_loss']

epochs=range(len(acc)) # Get number of epochs

#------------------------------------------------
# Plot training and validation accuracy per epoch
#------------------------------------------------
plt.plot(epochs, acc, 'r')
plt.plot(epochs, val_acc, 'b')
plt.title('Training and validation accuracy')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend(['Training', 'Validation'])
plt.show()
print("")

#------------------------------------------------
# Plot training and validation loss per epoch
#------------------------------------------------
plt.plot(epochs, loss, 'r')
plt.plot(epochs, val_loss, 'b')
plt.title('Training and validation loss')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend(['Training', 'Validation'])
plt.show()
```

# Model Predictions

```python
predictions = (model.predict(x_val))

predictions = np.ravel(predictions)
rounded_predictions = []
for i in predictions:
    rounded_predictions.append(round(i))
correct = 0

for i in range(0, len(rounded_predictions)):
    if (y_val[i]) == rounded_predictions[i]:
        correct+=1

print("Prediction\t", "Actual")
for i in range(0, len(rounded_predictions)):
    print("\t"+str(rounded_predictions[i])+"\t", y_val[i])
print("Validation Accuracy =",(correct / len(rounded_predictions)) * 100)
```
